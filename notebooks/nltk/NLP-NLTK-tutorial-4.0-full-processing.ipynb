```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
```

```python
# not important word that available in text (noise)
# can be use to remove it from the text
stopwords.words('english')
```

```python
sentences = nltk.sent_tokenize(paragraph) # save tokenized sentences
```

```python
type(sentences) # tokenize the paragraph
```

## Application: PorterStemmer
1. Make it into a list
2. Apply stopwords
3. Apply stemming

```python
from nltk.stem import PorterStemmer
```

```python
stemmer = PorterStemmer()
```

```python
# Apply Stopwords, Filter, Stemming
for i in range(len(sentences)):
  words = nltk.word_tokenize(sentences[i]) # get list of word from the sentences
  words = [stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))] # at each word if not present in stopword, apply stemming
  sentences[i] = ' '.join(words) # convert list & join all the separated words into sentences
```

## Application: SnowballStemmer

```python
from nltk.stem import SnowballStemmer
snowballstemmer = SnowballStemmer('english')
```

```python
# Apply Stopwords, Filter, Stemming
for i in range(len(sentences)):
  words = nltk.word_tokenize(sentences[i]) # get list of word from the sentences
  words = [snowballstemmer.stem(word) for word in words if word not in set(stopwords.words('english'))] # at each word if not present in stopword, apply stemming
  sentences[i] = ' '.join(words) # convert list & join all the separated words into sentences
```

```python
sentences
```

## Application: WordNetLemmatizer

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
# Apply Stopwords, Filter, Stemming
for i in range(len(sentences)):
  words = nltk.word_tokenize(sentences[i]) # get list of word from the sentences
  words = [lemmatizer.lemmatize(word.lower()) for word in words if word not in set(stopwords.words('english'))] # at each word if not present in stopword, apply stemming
  sentences[i] = ' '.join(words) # convert list & join all the separated words into sentences
```
