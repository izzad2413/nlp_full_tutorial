## Named Entity Recognition

```python
sentence = "The Eiffel Tower was built from 1887 to 1889 by Gustave Eiffel, whose company specialized in building metal frameworks and structures."
```

```python
import nltk
```

```python
nltk.download('punkt') # nltk.sent_tokenize
nltk.download('averaged_perceptron_tagger') # nltk.pos_tag
nltk.download('maxent_ne_chunker') # nltk.ne_chunk
nltk.download('words') # nltk.ne_chunk
```

## Tokenize

```python
words = nltk.sent_tokenize(sentence) # save tokenized sentences
print(f'Paragraph has been tokenize into {type(words)}')
```

```python
words
```

## Part of Speech

```python
tag_elements = nltk.pos_tag(words)
```

```python
tag_elements
```

## Named Entity Recognition

```python
nltk.ne_chunk(tag_elements).draw()
```

```python
chunked_tree = nltk.ne_chunk(tag_elements)
with open('chunked_tree.txt', 'w') as f:
    f.write(chunked_tree.pformat())
```
